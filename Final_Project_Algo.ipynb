```python
import random
import os
from IPython.display import Image, display
#the win rate calculate by win/total match for each team
PACI = {'PRX':0.75,'DRX':0.66,'T1':0.52,'GEN':0.41,'TS':0.5,'ZETA':0.47,'RRQ':0.38,'GES':0.27,'TLN':0.42,'DFM':0,'BLD':0.83}
EMEA = {'FNC':0.55,'TL':0.55,'FUT':0.5,'NAVI':0.52,'GIA':0.52,'KOI':0.28,'TH':0.18,'KC':0.28,'VIT':0.38,'BBL':0.27,'GM':0.75}
AMER = {'EG':0.62,'LLL':0.74,'NRG':0.61,'C9':0.64,'100':0.46,'SEN':0.38,'KRU':0.29,'MIBR':0.27,'FUR':0.4,'LEV':0.44,'G2':0.62}
#chinese team not be annouce yet, so I use 2022 participation team to predict
CHIN = {'EDG':0.48,'BLG':0.4,'FPX':0.52,'ASE':0.76,'TYL':0.4,'TCE':0.5,'ME':0.33,'WB':0.37,'FAM':0,'RNG':0,'NTER':0.33,'DRG':0.37}

TEAM_PIC = {'PRX':'/content/PAC/PRX.png','DRX':'/content/PAC/DRX.png','T1':'/content/PAC/T1.png','GEN':'/content/PAC/GEN.png','TS':'/content/PAC/TS.png','ZETA':'/content/PAC/ZETA.png','RRQ':'/content/PAC/RRQ.png','GES':'/content/PAC/GES.png','TLN':'/content/PAC/TLN.png','DFM':'/content/PAC/DFM.png','BLD':'/content/PAC/BLD.png'
,'FNC':'/content/EMEA/FNC.png','TL':'/content/EMEA/TL.png','FUT':'/content/EMEA/FUT.png','NAVI':'/content/EMEA/NAVI.png','GIA':'/content/EMEA/GIA.png','KOI':'/content/EMEA/KOI.png','TH':'/content/EMEA/TH.png','KC':'/content/EMEA/KC.png','VIT':'/content/EMEA/VIT.png','BBL':'/content/EMEA/BBL.png','GM':'/content/EMEA/GM.png'
,'EG':'/content/AMER/EG.png','LLL':'/content/AMER/LLL.png','NRG':'/content/AMER/NRG.png','C9':'/content/AMER/C9.png','100':'/content/AMER/100.png','SEN':'/content/AMER/SEN.png','KRU':'/content/AMER/KRU.png','MIBR':'/content/AMER/MIBR.png','FUR':'/content/AMER/FUR.png','LEV':'/content/AMER/LEV.png','G2':'/content/AMER/G2.png'
,'EDG':'/content/CN/EDG.png','BLG':'/content/CN/BLG.png','FPX':'/content/CN/FPX.png','ASE':'/content/CN/ASE.png','TYL':'/content/CN/TYL.png','TCE':'/content/CN/TCE.png','ME':'/content/CN/ME.png','WB':'/content/CN/WB.png','FAM':'/content/CN/FAM.png','RNG':'/content/CN/RNG.png','NTER':'/content/CN/NTER.png','DRG':'/content/CN/DRG.png'}

#Store the winning time for each team
PACI_WIN = {'PRX':0,'DRX':0,'T1':0,'GEN':0,'TS':0,'ZETA':0,'RRQ':0,'GES':0,'TLN':0,'DFM':0,'BLD':0}
EMEA_WIN = {'FNC':0,'TL':0,'FUT':0,'NAVI':0,'GIA':0,'KOI':0,'TH':0,'KC':0,'VIT':0,'BBL':0,'GM':0}
AMER_WIN = {'EG':0,'LLL':0,'NRG':0,'C9':0,'100':0,'SEN':0,'KRU':0,'MIBR':0,'FUR':0,'LEV':0,'G2':0}
CHIN_WIN = {'EDG':0,'BLG':0,'FPX':0,'ASE':0,'TYL':0,'TCE':0,'ME':0,'WB':0,'FAM':0,'RNG':0,'NTER':0,'DRG':0}
#Function to predict groupstage result
def ch_predict(team1,team2):
  win_rate1 = CHIN[team1]
  win_rate2 = CHIN[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2

def paci_predict(team1,team2):
  win_rate1 = PACI[team1]
  win_rate2 = PACI[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2

def emea_predict(team1,team2):
  win_rate1 = EMEA[team1]
  win_rate2 = EMEA[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2
def amer_predict(team1,team2):
  win_rate1 = AMER[team1]
  win_rate2 = AMER[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2


chin_group = []
paci_group = []
emea_group = []
amer_group = []
MASTER_MADRID = {}
MASTER_SHANGHAI = {}
master_shanghai_playoff = {}
master_shanghai_group = {}
shanghai_a_win = {}
shanghai_b_win = {}
VCT_CHAMPION = {}
champion_a_win = {}
champion_b_win = {}
champion_c_win = {}
champion_d_win = {}
vct_champion_playoff = {}



#loop to find winner for group stage
for team1 in CHIN.keys():
  for team2 in CHIN.keys():
    if team1 != team2:
      winner = ch_predict(team1,team2)
      CHIN_WIN[winner] +=1
      chin_group.append((team1,team2, winner))

for team1 in PACI.keys():
  for team2 in PACI.keys():
    if team1 != team2:
      winner = paci_predict(team1,team2)
      PACI_WIN[winner] += 1
      paci_group.append((team1,team2, winner))

for team1 in EMEA.keys():
  for team2 in EMEA.keys():
    if team1 != team2:
      winner = emea_predict(team1,team2)
      EMEA_WIN[winner] +=1
      emea_group.append((team1,team2, winner))

for team1 in AMER.keys():
  for team2 in AMER.keys():
    if team1 != team2:
      winner = amer_predict(team1,team2)
      AMER_WIN[winner] += 1
      amer_group.append((team1,team2, winner))
#Sorting the most winning to low
chin_table = sorted(CHIN_WIN.items(), key=lambda x: x[1], reverse=True)
paci_table = sorted(PACI_WIN.items(), key=lambda x: x[1], reverse=True)
emea_table = sorted(EMEA_WIN.items(), key=lambda x: x[1], reverse=True)
amer_table = sorted(AMER_WIN.items(), key=lambda x: x[1], reverse=True)

start1 = input("Type something to start the predict:")
print("")
print("PACIFIC LEAGUE")
print(paci_group)
print(PACI_WIN)
print(paci_table)
print(" ")
print("EMEA LEAGUE")
print(emea_group)
print(EMEA_WIN)
print(emea_table)
print(" ")
print("AMERICA LEAGUE")
print(amer_group)
print(AMER_WIN)
print(amer_table)
print(" ")
print("CHINA LEAGUE")
print(chin_group)
print(CHIN_WIN)
print(chin_table)



#add top 2 team form each league into Master Madrid
for key, value in chin_table[:2]:
    MASTER_MADRID[key] = value


for key, value in paci_table[:2]:
    MASTER_MADRID[key] = value


for key, value in emea_table[:2]:
    MASTER_MADRID[key] = value


for key, value in amer_table[:2]:
    MASTER_MADRID[key] = value

#Change winning time back to it win rate
for key, value in chin_table[:2]:
    MASTER_MADRID[key] = CHIN[key]

for key, value in paci_table[:2]:
    MASTER_MADRID[key] = PACI[key]

for key, value in emea_table[:2]:
    MASTER_MADRID[key] = EMEA[key]

for key, value in amer_table[:2]:
    MASTER_MADRID[key] = AMER[key]
print("")
start2 = input("Type something to start the predict:")
print(" ")
print("VCT Master Madrid Participation Team")
MADRID_PIC = {team: picture_path for team, picture_path in TEAM_PIC.items() if team in MASTER_MADRID}
for team, picture_path in MADRID_PIC.items():
    print(f"Team: {team}")
    display(Image(filename=picture_path))
print(" ")
print("VCT Master Madrid:")


def master_madrid_match(teams):
    random.shuffle(teams)
    madrid_matches = []

    # Pair teams for each match
    for i in range(0, len(teams), 2):
        match = (teams[i], teams[i+1])
        madrid_matches.append(match)

    return madrid_matches

# Function to simulate a playoff stage
def madrid_playoff_stage(teams, win_rates):
    madrid_playoff = master_madrid_match(teams)
    winners = []

    for i, match in enumerate(madrid_playoff, 1):
        team1, team2 = match
        win_rate1 = win_rates[team1]
        win_rate2 = win_rates[team2]

        total = win_rate1 + win_rate2
        predict = random.uniform(0, total)
        winner = team1 if predict <= win_rate1 else team2
        winners.append(winner)
        print(f"Playoff 1 Match:{team1} vs {team2} => Winner: {winner}")
        print("")


    while len(winners) > 1:
        madrid_playoff = master_madrid_match(winners)
        winners = []

        for i, match in enumerate(madrid_playoff, 1):
            team1, team2 = match
            win_rate1 = win_rates[team1]
            win_rate2 = win_rates[team2]

            total = win_rate1 + win_rate2
            predict = random.uniform(0, total)
            winner = team1 if predict <= win_rate1 else team2
            winners.append(winner)
            print(f"Playoff 2 Match:{team1} vs {team2} => Winner: {winner}")
            print(" ")

    return winners[0]

madrid_champion = None
madrid_champion = madrid_playoff_stage(list(MASTER_MADRID.keys()), MASTER_MADRID)
print(f"VCT Master Madrid Champion is: {madrid_champion}")

if madrid_champion:
    picture_path = MADRID_PIC.get(madrid_champion)
    if picture_path:
        display(Image(filename=picture_path))

#-----------------------------------------------------------------------------------------------
#Add team to master shanghai
for key, value in chin_table[:3]:
    MASTER_SHANGHAI[key] = CHIN[key]

for key, value in paci_table[:3]:
    MASTER_SHANGHAI[key] = PACI[key]

for key, value in emea_table[:3]:
    MASTER_SHANGHAI[key] = EMEA[key]

for key, value in amer_table[:3]:
    MASTER_SHANGHAI[key] = AMER[key]
#2nd and 3rd seed from each league enter group stage
for key, value in chin_table[1:3]:
    master_shanghai_group[key] = CHIN[key]

for key, value in paci_table[1:3]:
    master_shanghai_group[key] = PACI[key]

for key, value in emea_table[1:3]:
    master_shanghai_group[key] = EMEA[key]

for key, value in amer_table[1:3]:
    master_shanghai_group[key] = AMER[key]
#First seed from every league enter the playoff stage directly
for key, value in chin_table[:1]:
    master_shanghai_playoff[key] = CHIN[key]

for key, value in paci_table[:1]:
    master_shanghai_playoff[key] = PACI[key]

for key, value in emea_table[:1]:
    master_shanghai_playoff[key] = EMEA[key]

for key, value in amer_table[:1]:
    master_shanghai_playoff[key] = AMER[key]

def master_shaghai_group_divide(teams):
  master2_group = list(teams.items())
  random.shuffle(master2_group)
  group_a = {}
  group_b = {}
  divide_group = len(master2_group) // 2


  for i, (key, value) in enumerate(master2_group):
    if i < divide_group:
      group_a[key] = value
    else:
      group_b[key] = value


  return group_a, group_b

def shanghai_a_predict(team1,team2):
  win_rate1 = shanghai_group_a[team1]
  win_rate2 = shanghai_group_a[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2

def shanghai_b_predict(team1,team2):
  win_rate1 = shanghai_group_b[team1]
  win_rate2 = shanghai_group_b[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2

shanghai_a_group = []
shanghai_b_group = []

shanghai_group_a,shanghai_group_b = master_shaghai_group_divide(master_shanghai_group)
shanghai_a_win = {key: 0 for key in shanghai_group_a}
shanghai_b_win = {key: 0 for key in shanghai_group_b}

for team1 in shanghai_group_a.keys():
  for team2 in shanghai_group_a.keys():
    if team1 != team2:
      master2_winner = shanghai_a_predict(team1,team2)
      shanghai_a_win[master2_winner] += 1
      shanghai_a_group.append((team1,team2, master2_winner))

for team1 in shanghai_group_b.keys():
  for team2 in shanghai_group_b.keys():
    if team1 != team2:
      master2_winner = shanghai_b_predict(team1,team2)
      shanghai_b_win[master2_winner] +=1
      shanghai_b_group.append((team1,team2, master2_winner))

shanghai_a_table = sorted(shanghai_a_win.items(), key=lambda x: x[1], reverse=True)
shanghai_b_table = sorted(shanghai_b_win.items(), key=lambda x: x[1], reverse=True)

for key, value in shanghai_a_table[:2]:
    master_shanghai_playoff[key] = MASTER_SHANGHAI[key]

for key, value in shanghai_b_table[:2]:
    master_shanghai_playoff[key] = MASTER_SHANGHAI[key]



def master_shanghai_match(teams):
    random.shuffle(teams)
    shanghai_matches = []

    # Pair team
    for i in range(0, len(teams), 2):
        match = (teams[i], teams[i+1])
        shanghai_matches.append(match)

    return shanghai_matches
print("")
start3 = input("Type something to start the predict:")
print("")
print("VCT Master Shanghai Participation Team")
print("")
SHANGHAI_PIC = {team: picture_path for team, picture_path in TEAM_PIC.items() if team in MASTER_SHANGHAI}
for team, picture_path in SHANGHAI_PIC.items():
    print(f"Team: {team}")
    display(Image(filename=picture_path))
print("")
print("VCT Master Shanghai Group Stage:")
print("")
print("VCT Master Shanghai Group A")
print("")
print(shanghai_a_table)
print("")
print("VCT Master Shanghai Group B")
print("")
print(shanghai_b_table)
print("")
print("VCT Master Shanghai Playoff Stage")
print("")

# Function to simulate a playoff stage
def shanghai_playoff_stage(teams, win_rates):

  shanghai_playoff = master_shanghai_match(teams)
  winners = []

  for i, match in enumerate(shanghai_playoff, 1):
    team1, team2 = match
    win_rate1 = win_rates[team1]
    win_rate2 = win_rates[team2]

    total = win_rate1 + win_rate2
    predict = random.uniform(0, total)
    winner = team1 if predict <= win_rate1 else team2
    winners.append(winner)
    print(f"Playoff 1 Match:{team1} vs {team2} => Winner: {winner}")
    print("")


  while len(winners) > 1:
    shanghai_playoff = master_shanghai_match(winners)
    winners = []

    for i, match in enumerate(shanghai_playoff, 1):
      team1, team2 = match
      win_rate1 = win_rates[team1]
      win_rate2 = win_rates[team2]

      total = win_rate1 + win_rate2
      predict = random.uniform(0, total)
      winner = team1 if predict <= win_rate1 else team2
      winners.append(winner)
      print(f"Playoff 2 Match:{team1} vs {team2} => Winner: {winner}")
      print(" ")

  return winners[0]


shanghai_champion = shanghai_playoff_stage(list(master_shanghai_playoff.keys()), master_shanghai_playoff)
print(f"VCT Master Shanghai Champion is: {shanghai_champion}")
if shanghai_champion:
    picture_path = TEAM_PIC.get(shanghai_champion)
    if picture_path:
        display(Image(filename=picture_path))
print("")
start4 = input("Type something to start the predict:")

#VCT Champion 2024
for key, value in chin_table[:4]:
    VCT_CHAMPION[key] = CHIN[key]

for key, value in paci_table[:4]:
    VCT_CHAMPION[key] = PACI[key]

for key, value in emea_table[:4]:
    VCT_CHAMPION[key] = EMEA[key]

for key, value in amer_table[:4]:
    VCT_CHAMPION[key] = AMER[key]

#Function divided 16 teams into 4 groups
def champion_group_divide(teams):
  champion_group = list(teams.items())
  random.shuffle(champion_group)
  group_a = {}
  group_b = {}
  group_c = {}
  group_d = {}
  divide_group = len(champion_group) // 4

  for i, (key, value) in enumerate(champion_group):
    if i < divide_group:
      group_a[key] = value
    elif i < 2 * divide_group:
      group_b[key] = value
    elif i < 3 * divide_group:
      group_c[key] = value
    else:
      group_d[key] = value

  return group_a, group_b,group_c,group_d

#Apply team in each group to winning dic for count number of win
champion_group_a,champion_group_b,champion_group_c,champion_group_d = champion_group_divide(VCT_CHAMPION)
champion_a_win = {key: 0 for key in champion_group_a}
champion_b_win = {key: 0 for key in champion_group_b}
champion_c_win = {key: 0 for key in champion_group_c}
champion_d_win = {key: 0 for key in champion_group_d}

champion_a_group = []
champion_b_group = []
champion_c_group = []
champion_d_group = []

#predict groupstage
def champion_a_predict(team1,team2):
  win_rate1 = champion_group_a[team1]
  win_rate2 = champion_group_a[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2
def champion_b_predict(team1,team2):
  win_rate1 = champion_group_b[team1]
  win_rate2 = champion_group_b[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2
def champion_c_predict(team1,team2):
  win_rate1 = champion_group_c[team1]
  win_rate2 = champion_group_c[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2
def champion_d_predict(team1,team2):
  win_rate1 = champion_group_d[team1]
  win_rate2 = champion_group_d[team2]
  total = win_rate1 + win_rate2
  predict = random.uniform(0, total)
  if predict <= win_rate1:
    return team1
  else:
    return team2

#Finding winner for each group
for team1 in champion_group_a.keys():
  for team2 in champion_group_a.keys():
    if team1 != team2:
      champion_group_winner = champion_a_predict(team1,team2)
      champion_a_win[champion_group_winner] += 1
      champion_a_group.append((team1,team2, champion_group_winner))

for team1 in champion_group_b.keys():
  for team2 in champion_group_b.keys():
    if team1 != team2:
      champion_group_winner = champion_b_predict(team1,team2)
      champion_b_win[champion_group_winner] += 1
      champion_b_group.append((team1,team2, champion_group_winner))

for team1 in champion_group_c.keys():
  for team2 in champion_group_c.keys():
    if team1 != team2:
      champion_group_winner = champion_c_predict(team1,team2)
      champion_c_win[champion_group_winner] += 1
      champion_c_group.append((team1,team2, champion_group_winner))

for team1 in champion_group_d.keys():
  for team2 in champion_group_d.keys():
    if team1 != team2:
      champion_group_winner = champion_d_predict(team1,team2)
      champion_d_win[champion_group_winner] += 1
      champion_d_group.append((team1,team2, champion_group_winner))

#Sorting from most to least winning
champion_a_table = sorted(champion_a_win.items(), key=lambda x: x[1], reverse=True)
champion_b_table = sorted(champion_b_win.items(), key=lambda x: x[1], reverse=True)
champion_c_table = sorted(champion_c_win.items(), key=lambda x: x[1], reverse=True)
champion_d_table = sorted(champion_d_win.items(), key=lambda x: x[1], reverse=True)

for key, value in champion_a_table[:2]:
    vct_champion_playoff[key] = VCT_CHAMPION[key]

for key, value in champion_b_table[:2]:
    vct_champion_playoff[key] = VCT_CHAMPION[key]

for key, value in champion_c_table[:2]:
    vct_champion_playoff[key] = VCT_CHAMPION[key]

for key, value in champion_d_table[:2]:
    vct_champion_playoff[key] = VCT_CHAMPION[key]

#Function to random teams to facing each other
def vct_champion_match(teams):
    random.shuffle(teams)
    champion_matches = []

    # Pair teams for each match
    for i in range(0, len(teams), 2):
        match = (teams[i], teams[i+1])
        champion_matches.append(match)

    return champion_matches
print("")
print("VCT Champion Participation Team")
print("")
CHAMPION_PIC = {team: picture_path for team, picture_path in TEAM_PIC.items() if team in VCT_CHAMPION}
for team, picture_path in CHAMPION_PIC.items():
    print(f"Team: {team}")
    display(Image(filename=picture_path))
print("")
print("VCT Champion Group Stage:")
print("")
print("Champion Group A")
print("")
print(champion_a_table)
print("")
print("Champion Group B")
print("")
print(champion_b_table)
print("")
print("Champion Group C")
print("")
print(champion_c_table)
print("")
print("Champion Group D")
print("")
print(champion_d_table)
print("")
print("VCT Champion Playoff Stage")
print("")

#Simulated Playoff Stage
def champion_playoff_stage(teams, win_rates):

  champion_playoff = vct_champion_match(teams)
  winners = []

  for i, match in enumerate(champion_playoff, 1):
    team1, team2 = match
    win_rate1 = win_rates[team1]
    win_rate2 = win_rates[team2]

    total = win_rate1 + win_rate2
    predict = random.uniform(0, total)
    winner = team1 if predict <= win_rate1 else team2
    winners.append(winner)
    print(f"Playoff 1 Match:{team1} vs {team2} => Winner: {winner}")
    print("")


  while len(winners) > 1:
    champion_playoff = vct_champion_match(winners)
    winners = []

    for i, match in enumerate(champion_playoff, 1):
      team1, team2 = match
      win_rate1 = win_rates[team1]
      win_rate2 = win_rates[team2]

      total = win_rate1 + win_rate2
      predict = random.uniform(0, total)
      winner = team1 if predict <= win_rate1 else team2
      winners.append(winner)
      print(f"Playoff 2 Match:{team1} vs {team2} => Winner: {winner}")
      print(" ")

  return winners[0]

vct_2024_champion = champion_playoff_stage(list(vct_champion_playoff.keys()), vct_champion_playoff)
print(f"VCT 2024 Champion is: {vct_2024_champion}")
if vct_2024_champion:
    picture_path = TEAM_PIC.get(vct_2024_champion)
    if picture_path:
        display(Image(filename=picture_path))
print("")
end1 = input("Type something to see the all champion result:")
print("")
print("All VCT 2024 Tournament Champion")
print("")
print(f"VCT Master 1 Madrid Champion : {madrid_champion}")
if madrid_champion:
    picture_path = TEAM_PIC.get(madrid_champion)
    if picture_path:
        display(Image(filename=picture_path))
print("")
print(f"VCT Master 2 Shanghai Champion: {shanghai_champion}")
if shanghai_champion:
    picture_path = TEAM_PIC.get(shanghai_champion)
    if picture_path:
        display(Image(filename=picture_path))
print("")
print(f"VCT 2024 World Champion : {vct_2024_champion}")
if vct_2024_champion:
    picture_path = TEAM_PIC.get(vct_2024_champion)
    if picture_path:
        display(Image(filename=picture_path))
```
